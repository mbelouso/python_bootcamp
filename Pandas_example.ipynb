```python
# Code to import a CSV file into a pandas DataFrame and then plot the data using matplotlib

import pandas as pd
import matplotlib.pyplot as plt 
from scipy import stats
import numpy as np
# Load the CSV file into a pandas DataFrame
df = pd.read_csv('data.csv')

# Display the first few rows of the DataFrame
print(df.head())

# Print the DataFrame's shape
print("Shape of DataFrame:", df.shape)

# Print the Dataframs column identifiers
print("Column identifiers:", df.columns)

# Print the the [Temperature] column from the DataFrame
print("Temperature column:")
print(df['Temperature'])
```

```python
# Use Numpy to calculate the mean and standard deviation of the 'Temperature' column
mean_temp = np.mean(df['Temperature'])
std_temp = np.std(df['Temperature'])

# Print the mean and standard deviation
print("Mean Temperature:", mean_temp)
print("Standard Deviation of Temperature:", std_temp)

# Use SciPy to calculate the correlation between 'Temperature' and 'Humidity'
correlation = stats.pearsonr(df['Temperature'], df['Humidity'])
print("Correlation between Temperature and Humidity:", correlation[0])
```

```python
# Plot the data as a scatter plot
plt.figure(figsize=(10, 5))
plt.subplot(1,2,1)

plt.scatter(df['Date'], df['Temperature'], color='blue', label='Temperature')
plt.scatter(df['Date'], df['Humidity'], color='green', label='Humidity')
plt.xlabel('Date')
# set separate y-axis for clarity
plt.ylabel('Temperature / Humidity')
plt.title('Temperature and Humidity Over Time')
plt.legend()
plt.xticks(rotation=45)

plt.subplot(1,2,2)
plt.hist(df['Temperature'], bins=5, color='blue', alpha=0.5, label='Temperature')
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.title('Histogram of Temperature')
plt.legend()

plt.tight_layout()
plt.show()
```

# Now plot Temp vs Humidity and plot linear regression....

```python
# Perform a linear regression analysis on the 'Temperature' and 'Humidity' columns

slope, intercept, r_value, p_value, std_err = stats.linregress(df['Temperature'], df['Humidity'])
print(f"Linear Regression: y = {slope:.2f}x + {intercept:.2f}, R-squared = {r_value**2:.2f}")

# Plot Temp vs Humidity as a Scatter plot
plt.figure(figsize=(8, 6))
plt.scatter(df['Temperature'], df['Humidity'], color='purple', alpha=0.6)
# plot the regression line on top of the scatter plot
plt.plot(df['Temperature'], slope * df['Temperature'] + intercept, color='red', label='Regression Line')
plt.title('Temperature vs Humidity')
plt.xlabel('Temperature')
plt.ylabel('Humidity')
plt.grid(True)
plt.show()
# Test
```
