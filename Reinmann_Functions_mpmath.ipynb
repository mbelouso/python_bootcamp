```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.special import zeta

# Define a range of x values (real part)
x = np.linspace(0.1, 10, 500)

# Compute the Riemann zeta function for each x
zeta_values = zeta(x)

# Plotting
plt.figure(figsize=(8, 6))
plt.plot(x, zeta_values.real, label='Real part of Zeta(x)', color='blue')
plt.plot(x, zeta_values.imag, label='Imaginary part of Zeta(x)', color='red', linestyle='--')

# Add labels and title
plt.title("Plot of the Riemann Zeta Function", fontsize=14)
plt.xlabel("x", fontsize=12)
plt.ylabel("Zeta(x)", fontsize=12)
plt.legend()

# Display the plot
plt.grid(True)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from mpmath import zeta

# Define the grid of complex numbers (Real and Imaginary parts)
real = np.linspace(0.1, 4, 50)  # Real part range
imag = np.linspace(-4, 4, 50)  # Imaginary part range

# Create a meshgrid of complex numbers
real_grid, imag_grid = np.meshgrid(real, imag)
complex_grid = real_grid + 1j * imag_grid

# Initialize array for storing zeta values
zeta_values = np.zeros_like(complex_grid, dtype=complex)

# Compute the Riemann zeta function for each complex number
for i in range(complex_grid.shape[0]):
    for j in range(complex_grid.shape[1]):
        zeta_values[i, j] = zeta(complex_grid[i, j])  # Use zetafn for individual complex numbers

# Plotting the magnitude of the zeta function on the complex plane
plt.figure(figsize=(8, 6))
plt.contourf(real_grid, imag_grid, np.abs(zeta_values), levels=50, cmap='viridis')
plt.colorbar(label='Magnitude of Zeta(s)')

# Add labels and title
plt.title("Riemann Zeta Function on the Complex Plane", fontsize=14)
plt.xlabel("Real Part of s", fontsize=12)
plt.ylabel("Imaginary Part of s", fontsize=12)

# Display the plot
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from mpmath import zeta

# Define the grid of complex numbers (Real and Imaginary parts)
real = np.linspace(0.1, 5, 50)  # Real part range (adjust density as needed)
imag = np.linspace(-5, 5, 50)   # Imaginary part range (adjust density as needed)

# Create a meshgrid of complex numbers
real_grid, imag_grid = np.meshgrid(real, imag)
complex_grid = real_grid + 1j * imag_grid

# Initialize arrays for the real and imaginary parts of the zeta function
zeta_real = np.zeros_like(complex_grid, dtype=float)
zeta_imag = np.zeros_like(complex_grid, dtype=float)

# Compute the Riemann zeta function for each complex number
for i in range(complex_grid.shape[0]):
    for j in range(complex_grid.shape[1]):
        zeta_value = zeta(complex_grid[i, j])  # Compute the zeta function at the complex point
        zeta_real[i, j] = zeta_value.real
        zeta_imag[i, j] = zeta_value.imag

# Plotting the complex plane and the transformation vectors
plt.figure(figsize=(8, 8))

# Plot the original complex plane (input points)
plt.quiver(real_grid, imag_grid, zeta_real, zeta_imag, angles='xy', scale_units='xy', scale=0.5, color='blue')

# Adding labels and title
plt.title("Riemann Zeta Function Transformation on the Complex Plane", fontsize=14)
plt.xlabel("Real Part of s", fontsize=12)
plt.ylabel("Imaginary Part of s", fontsize=12)

# Set equal scaling for both axes
plt.axis('equal')

# Display the plot
plt.grid(True)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from mpmath import zetazero

# Number of non-trivial zeros to plot
num_zeros = 20

# Compute the first `num_zeros` non-trivial zeros of the Riemann zeta function
zeros = [zetazero(n) for n in range(1, num_zeros + 1)]

# Extract the real and imaginary parts of the zeros
real_zeros = [z.real for z in zeros]
imag_zeros = [z.imag for z in zeros]

# Plot the non-trivial zeros on the complex plane
plt.figure(figsize=(8, 6))
plt.scatter(real_zeros, imag_zeros, color='red', label='Non-trivial Zeros', zorder=5)

# Plot the critical line (Re(s) = 1/2)
plt.axvline(x=0.5, color='blue', linestyle='--', label='Critical Line (Re(s) = 1/2)')

# Add labels and title
plt.title("Non-Trivial Zeros of the Riemann Zeta Function", fontsize=14)
plt.xlabel("Real Part of s", fontsize=12)
plt.ylabel("Imaginary Part of s", fontsize=12)

# Set equal scaling for both axes
#plt.axis('equal')

# Display the plot
plt.legend()
plt.grid(True)
plt.show()
```
