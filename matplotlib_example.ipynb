```python
import matplotlib.pyplot as plt

x = [x for x in range(10)]
y = [x**2 for x in x]

print(x)
print(y)

# Plot 
plt.figure(figsize=(8, 6))
plt.scatter(x, y, color='red')


# Add labels and title
plt.title("Quadratic Function", fontsize=14)
plt.xlabel("X", fontsize=12)
plt.ylabel("Y", fontsize=12)

# Set equal scaling for both axes
#plt.axis('equal')

# Display the plot
plt.grid(True)
plt.savefig("Quadratic.png")
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(0, 10, 1000) # note linspace (start,stop,number_of_points)
print(x)

y = [x**2 for x in x]
print(y)

# Plot 
plt.figure(figsize=(8, 6))
plt.plot(x, y, color='red',linewidth=1.0)     


# Add labels and title
plt.title("Quadratic Function", fontsize=14)
plt.xlabel("X", fontsize=12)
plt.ylabel("Y", fontsize=12)

# Display the plot
plt.grid(True)
plt.savefig("Quadratic_continuous.png")
plt.show()
```
